```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
reg = linear_model.LinearRegression()
np.random.seed(19170030) #Номер зачетки
A = []
B = []
T = []
N = 10000
n = 1000
```

```python
for i in range(n):
  t = time.time()
  x = np.random.uniform(low = 150, high = 200, size = N)
  e = np.random.normal(loc = 0, scale = 20, size = N)
  y = -100 + x + e
  reg.fit(x.reshape(-1, 1), y.reshape(-1, 1))
  A.append(reg.intercept_[0])
  B.append(reg.coef_[0][0])
  T.append(time.time() - t)
```

```python
plt.figure(figsize = (10, 6))
plt.hist(B, bins = 20, edgecolor = "black", color = "white")
plt.title("Histogram of B estimates")
plt.show()
```
